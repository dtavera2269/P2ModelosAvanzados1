## **Data Cleaning**

Se importa la libreria **re** de Python que proporciona operaciones de coincidencia en expresiones regulares. Es decir, que verigivan si una cadena particular coincide con un expresión regular dada. Aqui se uso principalmente para la limpieza de los plots, quitando contracciones informales de palabras. Además, de eliminar signos de puntuación o carácteres especiales. 

```python
import re
def arregla_doble_espacio(sentence):
    cleaned_sentence = re.sub(r" '","'", sentence)
    return cleaned_sentence

def arregla_espacio2(sentence):
    cleaned_sentence = re.sub(r"' ","'", sentence)
    return cleaned_sentence

def arregla_abrev(sentence):
    cleaned_sentence = re.sub(r"won't"," will not", sentence)
    cleaned_sentence = re.sub(r"can\'t"," can not", cleaned_sentence)
    cleaned_sentence = re.sub(r"n\'t"," not", cleaned_sentence)
    cleaned_sentence = re.sub(r"\'re"," are", cleaned_sentence)
    cleaned_sentence = re.sub(r"\'s"," is", cleaned_sentence)
    cleaned_sentence = re.sub(r"\'d"," would", cleaned_sentence)
    cleaned_sentence = re.sub(r"\'ll"," will", cleaned_sentence)
    cleaned_sentence = re.sub(r"\'t"," not", cleaned_sentence)
    cleaned_sentence = re.sub(r"\'ve"," have", cleaned_sentence)
    cleaned_sentence = re.sub(r"\'m"," am", cleaned_sentence)
    return cleaned_sentence

def remove_punctuation(sentence):
    cleaned_sentence = re.sub(r'[?|!|\'|"|#]', '', sentence)
    cleaned_sentence = re.sub(r'[,|.|;|:|(|)|{|}|\|/|<|>]|-', ' ', cleaned_sentence)
    cleaned_sentence = cleaned_sentence.replace("\n"," ")
    cleaned_sentence = re.sub(' +', ' ', cleaned_sentence)
    return cleaned_sentence
```

```python

dataTraining['plot'] = dataTraining['plot'].apply(arregla_doble_espacio)
dataTraining['plot'] = dataTraining['plot'].apply(arregla_espacio2)
dataTraining['plot'] = dataTraining['plot'].apply(arregla_abrev)
dataTraining['plot'] = dataTraining['plot'].apply(remove_punctuation)
```

```python
X = plots.tolist()

vect = CountVectorizer(stop_words='english',lowercase=True)
X_dtm = vect.fit_transform(X)
words = list(vect.vocabulary_.keys())
vocabulary = {x: idx + 1 for idx, x in enumerate(set(words))}
lp=list()
max_len=300
for p in X:
    lidx=list()
    #print(p)
    listapalabras=p.split()
    #print(listapalabras)
    for w in listapalabras:
      if w in vocabulary.keys():
        lidx.append(vocabulary[w])
    lp.append(lidx)
    #print(len(lidx))
from keras.preprocessing import sequence
X_pad = sequence.pad_sequences(lp, maxlen=max_len)
len(X_pad)
X_pad
```

Create a LSTM neural network to predict the rating of a movie

```python
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers.recurrent import LSTM
from keras.layers.core import Dense, Dropout
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
from livelossplot import PlotLossesKeras
from keras import backend as K

X_train, X_test, y_train, y_test = train_test_split(X_pad, y, test_size=0.33, random_state=42)
X_train.shape
```

Se encuentran los mejores parámentros para la construcción de la red neuronal LSTM 

```python
range_capas=[1,2,3,4,5,6]
range_nodos=[20,100,200] #100,200,300,500
range_outdim=[20,100,200] #100,200
range_activ=['relu','sigmoid','softmax']
range_dropout=[0.1,0.3,0.7] #,0.5,0.7
best_c=0
best_n=0
best_od=0
best_act=''
best_acc=0
#best_model=
for c in range_dropout:
    for n in range_nodos:
      for od in range_outdim:
        for a in range_activ:
          K.clear_session()
          model = Sequential()
          model.add(Embedding(len(vocabulary) + 1, output_dim=od, input_length=X_pad.shape[1]))
          model.add(LSTM(od))
          model.add(Dense(n, activation= a))
          model.add(Dropout(c))
          model.add(Dense(1, activation='sigmoid'))
          model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
          model.fit(X_train, y_train, validation_data=[X_test, y_test], epochs=5, verbose=2 ) 
          model.summary()
          y_pred = model.predict_classes(X_test)[:,0]
          act_acc=(y_pred == y_test).mean()
          if act_acc>best_acc:
            best_c=c
            best_n=n
            best_od=od
            best_act=a
            best_acc=act_acc
            
            print("c:",c," n:",n," od:",od," act:",a," auc:",act_acc)
```

Luego de encontrar los mejores párametros se tienen los siguientes resultados 

**c: 0.1  n: 20  od: 100  act: softmax  auc: 0.6304681504221028**

```python
from keras import layers
from keras import backend as K
K.clear_session()
model = Sequential()
model.add(Embedding(len(vocabulary) + 1, 100, input_length=X_pad.shape[1]))
model.add(LSTM(100))
model.add(Dense(20, activation='softmax'))
model.add(Dropout(0.1))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

model.summary() 
```

```python
model.fit(X_train, y_train, validation_data=[X_test, y_test], epochs=10, verbose=1,
          callbacks=[PlotLossesKeras()], batch_size=50 )
```

```python
y_pred = model.predict_classes(X_test)[:,0]

print((y_pred == y_test).mean()) #60.7
```
